# Create playlist
For each band, get the most popular two songs

Considerations:

```python
"""Get top 2 songs per band"""
bands = ['weezer', 'squarepusher', 'best coast', 'delta spirit', 'rac']

# Save all track data
all_track_data = []

for band in bands:
#     band = bands[0]
    # Get data for this band
    band_data = sp.search(band)

    # Select 'items'
    items = band_data['tracks']['items']

    for item in items:
        # Getting track data
        track_data = {}
        track_data['album'] = item['album']['name']
        track_data['name'] = item['name']
        track_data['id'] = item['id']
        artist = item['artists'][0]['name']
        if artist.lower() == band:
            track_data['artist'] = artist
        else:
            continue

        all_track_data.append(track_data)

track_df = pd.DataFrame(all_track_data)
track_df
```
